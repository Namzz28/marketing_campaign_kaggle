## Data Cleaning

```python
data.info()
```

```python
data.head()
```

```python
# remove the missing value
data = data.dropna()
print(len(data))
```

```python
data["Dt_Customer"] = pd.to_datetime(data["Dt_Customer"], format = "%d-%m-%Y")
```

```python
days = []
d1 = max(data["Dt_Customer"]).date()
for date in data["Dt_Customer"]:
    days.append(d1 - date.date())
```

```python
data = data.drop(["Dt_Customer"], axis=1)
```

```python
data["Customer_For"] = days
```

```python
data["Customer_For"] = data["Customer_For"].dt.days
```

```python
data.head()
```

```python
# Encoding Categorical Columns
data["Marital_Status"].value_counts()
```

```python
data["Education"].value_counts()
```

```python
# Feature Engineering
# Age
data["Age"] = 2024 - data["Year_Birth"]

# Total Spending
data["Spent"] = data["MntFishProducts"] + data["MntMeatProducts"] + data["MntFruits"] + data["MntSweetProducts"] + data["MntWines"] + data["MntGoldProds"]
# total children
data["Children"] = data["Kidhome"]+data["Teenhome"]

#deriving new feature from Marital Status
data["Living_with"] = data["Marital_Status"].replace({"Married" : "Partner", "Together" : "Partner", "Absurd":"Alone", "Widow":"Alone", "YOLO":"Alone", "Divorced":"Alone", "Single":"Alone"})

#total number of memebers in the family
data["Family_Size"] = data["Living_with"].replace({"Alone" : 1, "Partner": 2}) + data["Children"]

#
data["Is_Parent"] = np.where(data["Children"] > 0, 1, 0)

data["Education"] = data["Education"].replace({"Basic" : "UG", "2n Cycle" : "UG", "Graduation":"Graduate", "Master":"PG", "PhD":"PG"})


data = data.drop(["Marital_Status", "Year_Birth", "ID", "Z_CostContact", "Z_Revenue"], axis=1)
```

## Preprocessing

```python
## Encoding Categorical Features.
data.columns
```

```python
s = (data.dtypes == "object" )
s[s].index
```

```python
# Label Encoding
le = LabelEncoder()
for col in s[s].index:
    data[col] = le.fit_transform(data[col])
```

```python
data.head()
```

```python
data.columns
```

```python
## Scaling the DataSet
ds = data.copy()
cols_del = ['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5',
       'AcceptedCmp1', 'AcceptedCmp2', 'Complain', 'Response']
ds = ds.drop(cols_del, axis=1)
scaler =  StandardScaler()
scaled_ds = pd.DataFrame(scaler.fit_transform(ds), columns = ds.columns)
```

```python
scaled_ds.head()
```

```python
len(scaled_ds.columns)
```

## Dimenstionality Reduction

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 3)
pca.fit(scaled_ds)
PCA_ds = pd.DataFrame(pca.transform(scaled_ds), columns = ["col1", "col2", "col3"])
```

## Clustering 

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
```

```python
elbow = KElbowVisualizer(KMeans(), k = 10)
elbow.fit(PCA_ds)
elbow.show()
```

```python
ac = AgglomerativeClustering(n_clusters = 4)
predict_ac = ac.fit_predict(PCA_ds)
PCA_ds["Clusters"] = predict_ac 
data["Clusters"] = predict_ac
```

```python
PCA_ds.columns
```

```python
sns.color_palette("hsv", n_colors=len(PCA_ds['Clusters'].unique()))
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
palette = sns.color_palette("hsv", n_colors=len(PCA_ds['Clusters'].unique()))
colors = [palette[cluster] for cluster in PCA_ds['Clusters']]
# Scatter plot
scatter = ax.scatter(
    PCA_ds['col1'], PCA_ds['col2'] ,PCA_ds['col3'],
    c=colors, s=50, edgecolors='k', label='Clusters'
)
```
